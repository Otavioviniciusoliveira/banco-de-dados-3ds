```python
# Importa biblioteca para visualizao de graficos
import matplotlib.pyplot as plt

# Dados de exemplo
regioes = ['Norte', 'Sul', 'Leste', 'Oeste']
vendas = [20, 30, 25, 25]

# Grfico de Pizza
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.pie(vendas, labels=regioes, autopct='%.1f%%', startangle=140)
plt.title('Participao de Vendas por Regio')

# Grfico de Barras
plt.subplot(1, 2, 2)
plt.bar(regioes, vendas, color='skyblue')
plt.title('Comparao de Vendas por Regio')
plt.xlabel('Regio')
plt.ylabel('Vendas (%)')

plt.tight_layout()
plt.show()
```
